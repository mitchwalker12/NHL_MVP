# Cleaning Data

```python
import pandas as pd
```

```python
mvps = pd.read_csv("mvps.csv") # reading in the mvp csv file
```

```python
mvps.head()
```

```python
mvps = mvps[["Player", "Year", "Votes", "Vote%"]] #removing duplicate columns
```

```python
mvps.head()
```

```python
players = pd.read_csv("players.csv") # read in player data
```

```python
players.head()
```

```python
del players["Unnamed: 0"] # remove unnecessary columns
```

```python
players
```

```python
del players["Rk"] # remove unnecessary columns
```

```python
# removing * (hall of famer)
players["Player"] = players["Player"].str.replace("*","", regex=False)
```

```python
players.rename(columns={'Unnamed: 29': "Year"}, inplace=True)
```

```python
players.head()
```

```python
# players who played on multiple teams are listed multiple times
# combining multiples
players[players["Player"] == "Alexander Sulzer"]
```

```python
def single_team(df):
    '''Merging duplicate players to a single entry/row.'''
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"]=="TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_team)
```

```python
players
```

```python
players.index = players.index.droplevel() # drops the unnecessary index levels
```

```python
players.index = players.index.droplevel() # drops the unnecessary index levels
```

```python
players.head() # if you slide over to year, you will see that players only have one value per year
```

```python
players[players["Player"] == "Alexander Sulzer"] # if you slide over to year, you will see that players only have one value per year
```

# Combining the Player & MVP Data

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
# outer merge - not all players won mvp
```

```python
combined[combined["Votes"] > 0] # shows players that received mvp votes
```

```python
combined
```

```python
# replace NaN values with zeros
combined[["Votes", "Vote%"]] = combined[["Votes", "Vote%"]].fillna(0)
```

```python
combined
```

```python
teams = pd.read_csv("teams.csv") # reading in the team data
```

```python
teams
```

```python
teams = teams[~teams["W"].str.contains("Division")].copy() # removing rows that have 'division'
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False) # removing the * (playoff teams)
```

```python
sorted(teams["Team"].unique())
```

```python
nicknames = {}
with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name = line.replace("\n","").split(",") #replacing the new line character with nothing
        nicknames[abbrev] = name 
```

```python
combined["Team"] = combined["Tm"].map(nicknames) # mapping teams to the nickname file
```

```python
combined.head()
```

```python
stats = combined.merge(teams, how="outer",on=["Team", "Year"]) # for every row in combined, we add team data
```

```python
stats.fillna(0)
```

```python
del stats["Unnamed: 0"] # removing unnecessary columns
```

```python
stats.dtypes # check data types. They're all string. Need to conver numbers to integers/floats.
```

```python
stats = stats.apply(pd.to_numeric, errors='ignore') # goes through every column in stats and tries convert to number
```

```python
stats.dtypes
```

```python
stats.to_csv("player_mvp_stats.csv") # write to a new csv file
```

```python
highest_scoring = stats[stats["GP_x"] > 70].sort_values("PTS_x", ascending=False).head(10) #sort by scoring
```

```python
highest_scoring.head() # determinne who has scored the most in the single season over the last 10 years
```

```python
highest_scoring.plot.bar("Player", "PTS_x")
```

```python
# see who scored the most in a single year
highest_scoring_by_year = stats.groupby("Year").apply(lambda x: x.sort_values("PTS_x", ascending=False).head(1))
```

```python
highest_scoring_by_year.plot.bar("Year", "PTS_x")  # barplot to show scoring by year
```
